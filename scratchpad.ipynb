```python
n=100000
mu = np.arange(-5,5,2.5)
K = len(mu)
eps = np.random.randn(n,K)
z = mu.reshape(1,-1)+eps
softmax = np.exp(z)
softmax = softmax/softmax.sum(axis=1,keepdims=True)
softmax1 = softmax.mean(axis=0)

tau = 0.001
softmax3 = np.argmax(z/tau,axis=1).reshape(-1,1)
softmax3 = softmax3 == np.arange(K).reshape(1,-1)
softmax3 = softmax3.mean(axis=0)

softmax2 = np.exp(mu)/np.exp(mu).sum()
print ' '.join('%0.4f' % s for s in softmax1)
print ' '.join('%0.4f' % s for s in softmax2)
print ' '.join('%0.4f' % s for s in softmax3)
```
