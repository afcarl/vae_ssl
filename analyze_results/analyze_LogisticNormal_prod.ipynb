```python
def loadDataHDF5(data):
    if isinstance(data,h5py.File) or isinstance(data,h5py.Group):
        return {k:loadDataHDF5(v) for k,v in data.iteritems()}
    elif isinstance(data,h5py.Dataset):
        return data.value
    else:
        print 'unhandled datatype: %s' % type(data)

def loadHDF5(filepath):
    with h5py.File(filepath,'r') as f:
        return loadDataHDF5(f)
```

```python
def plotExperiment(results,
                   title = '',
                   keys = ['cost','bound','accuracy'],
                   funcs = [min,min,max],
                   yscales = ['linear','linear','linear'],
                   ylims = [(100,600),(100,600),(0,1)],
                   locs = ['upper right','upper right','lower right'],
                   label_prefix=None
                   ):
    plt.figure(1,figsize=(18,4))
    for i,(key,f,yscale,ylim,loc) in enumerate(zip(keys,funcs,yscales,ylims,locs)):
        #print i,key
        plt.subplot(1,3,i+1)
        if label_prefix == None:
            label = 'best %s = %0.2f' % (key,f(results[key]))
        else:
            label='%s: best %s = %0.2f' % (label_prefix,key,f(results[key]))
        plt.plot(results['epochs'],results[key],label=label)
        plt.legend(loc=loc)
        plt.yscale(yscale)
        plt.title('%s %s' % (title,key))
        if ylim is not None:
            plt.ylim(ylim)
```

```python
def reshape(x,batch_size=100):
    y = []
    n = len(x)/batch_size
    for i in range(n):
        y.append(x[batch_size*i:batch_size*(i+1)].reshape(1,batch_size,-1))
    y = np.vstack(y)
    return y

def plotPercentiles(x=None,y=None,color='b',alpha=0.2,label=None,linewidth=1.):
    if y==None:
        xvals = None
        yvals = x
    else:
        xvals = x
        yvals = y
    p05_norm = np.percentile(yvals,5,axis=1)
    p10_norm = np.percentile(yvals,10,axis=1)
    p25_norm = np.percentile(yvals,25,axis=1)
    p50_norm = np.percentile(yvals,50,axis=1)
    p75_norm = np.percentile(yvals,75,axis=1)
    p90_norm = np.percentile(yvals,90,axis=1)
    p95_norm = np.percentile(yvals,95,axis=1)
    
    if xvals is None:
        xvals = np.arange(len(p50_norm))

    plt.plot(xvals,p50_norm,color=color,label=label,linewidth=linewidth)
    #plt.fill_between(xvals,p05_norm,p95_norm,color=color,alpha=alpha)
    plt.fill_between(xvals,p10_norm,p90_norm,color=color,alpha=alpha)
    plt.fill_between(xvals,p25_norm,p75_norm,color=color,alpha=alpha)
```

# q(y*|x)

```python
def plotqyx(epochs,mu,logcov2,Y,color='b'):
    #mu = reshape(train['mu_Z2_L'])
    #logcov2 = reshape(train['logcov2_Z2_L'])
    mu = reshape(logcov2)
    logcov2 = reshape(logcov2)
    sigma = np.exp(0.5*logcov2)
    y = Y
    mu_y = np.sum(mu*y,axis=-1)
    sigma_y = np.sum(sigma*y,axis=-1)
    eps = np.random.randn(*mu.shape)
    Z2 = mu + eps*sigma
    alpha = np.exp(Z2)/np.exp(Z2).sum(axis=-1,keepdims=True)
    alpha_y = np.sum(alpha*y,axis=-1)
    plotPercentiles(epochs,alpha_y,label='alpha_y: mean=%0.2f' % alpha_y[:,-1].mean(),color='r',alpha=0.05)
    alpha_ny = alpha*(1-y)
    alpha_sorted = np.sort(alpha_ny,axis=-1)[:,:,:-1]
    for k in range(alpha_sorted.shape[-1]):
        vals = alpha_sorted[:,:,k]
        plotPercentiles(epochs,vals,label='%s: mean alpha=%0.2f'%(k,np.mean(vals[:,-1])),color=color,alpha=0.05,linewidth=0.5)
    #plt.yscale('log')
    plt.xlim(0,500)
    plt.ylim(0,1)
    plt.legend()
    plt.xlabel('epochs')
```

```python
plt.figure(1,figsize=(18,8))
plt.subplot(2,2,1)
plotqyx(train['epochs'],train['mu_Z2_L'],train['logcov2_Z2_L'],
        Y=(np.expand_dims(train['YL'],axis=-1) == np.arange(10).reshape(1,1,-1)).astype('float'))
plt.title('%s\ntrain: alpha L' % (modelkey))

plt.subplot(2,2,2)
plotqyx(train['epochs'],train['mu_Z2_U'],train['logcov2_Z2_U'],
        Y=(np.expand_dims(train['YU'],axis=-1) == np.arange(10).reshape(1,1,-1)).astype('float'),color='g')
plt.title('%s\ntrain: alpha U' % (modelkey))

plt.subplot(2,2,3)
plotqyx(valid['epochs'],valid['mu_Z2_L'],valid['logcov2_Z2_L'],Y=reshape(valid['Y']))
plt.title('%s\nvalid: alpha L' % (modelkey))

plt.subplot(2,2,4)
plotqyx(valid['epochs'],valid['mu_Z2_U'],valid['logcov2_Z2_U'],Y=reshape(valid['Y']),color='g')
plt.title('%s\nvalid: alpha U' % (modelkey))
plt.tight_layout()
```
