```python
%pylab inline

import matplotlib.pyplot as plt
import h5py
import os
import re
import seaborn
```

```python
def findLatestResults(dirpath,minepoch=10):
    #files = [f for f in os.listdir(dirpath) if 'h5' in f]
    files = os.listdir(dirpath)
    filenames = [f for f in files if 'final.h5' in f]
    #if len(filenames) > 0:
    #    return os.path.join(dirpath,filenames[0])
    intermediates = [f for f in files if 'stats.h5' in f]
    if len(intermediates) > 0:
        intermediates = [[f,int(re.search('EP(\d+)-stats.h5',f).group(1))] for f in intermediates]
        intermediates = sorted(intermediates,key=lambda x:x[1],reverse=True)
        #for f in intermediates:
        #    print f
        latest_save,latest_epoch = intermediates[0]
        returnpath = os.path.join(dirpath,latest_save)
        if latest_epoch < minepoch:
            return None
        else:
            return returnpath
    else:
        return None
    
def getResults(filepath):
    results = {}
    with h5py.File(filepath,'r') as f:
        for k,v in f.iteritems():
            #results[k] = np.array(v)
            a = np.array(v)
            if len(a.shape) > 1:
                results[k] = {'epochs':a[:,0],'vals':a[:,1]}
            else:
                results[k] = a
    return results

def getResultsAcrossSeeds(dirpath,expt_name):
    resultsAcrossSeeds = []
    for s in os.listdir(os.path.join(dirpath,expt_name)):
        expt_seed_path = os.path.join(dirpath,expt_name,s)
        filepath = findLatestResults(expt_seed_path)
        if filepath == None:
            print '\n>>>WARNING: no results for %s' % s
        else:
            seed_results = getResults(filepath)
            resultsAcrossSeeds.append(seed_results)
    if len(resultsAcrossSeeds)>0:
        keys = resultsAcrossSeeds[0].keys()
        if isinstance(resultsAcrossSeeds[0][keys[0]],dict):
            # minimum length of epochs across seeds
            epochs = {k:min([len(r[k]['epochs']) for r in resultsAcrossSeeds]) for k in keys}
            # set epochs = {key:epoch_array}
            epochs = {k:resultsAcrossSeeds[0][k]['epochs'][:epochs[k]] for k in keys}
            #results = {k:{'epochs':resultsAcrossSeeds[0][k]['epochs'],'vals':np.array([r[k]['vals'] for r in resultsAcrossSeeds])} for k in keys}
            results = {k:{'epochs':epochs[k],'vals':np.array([r[k]['vals'][:len(epochs[k])] for r in resultsAcrossSeeds])} for k in keys}
        else:
            results = {k:np.array([r[k] for r in resultsAcrossSeeds]) for k in keys}
        return results
    else:
        return None

def getStats(x):
    stats = {}
    #print x.keys()
    for k,v in x.iteritems():
        stat = {}
        if isinstance(v,dict):
            stat['epochs'] = v['epochs']
            vals = v['vals']
        else:
            vals = v
        #raise NameError('thing')
        if len(vals.shape) > 1:
            stat['mean'] = vals.mean(axis=0)
            stat['p10'] = np.percentile(vals,10,axis=0)
            stat['p90'] = np.percentile(vals,90,axis=0)
        else:
            stat['mean'] = vals.mean(axis=0)
            stat['p10'] = np.percentile(vals,10)
            stat['p90'] = np.percentile(vals,90)
        stats[k] = stat
    return stats
        
```

```python
rootdir = '/scratch/jmj/theanomodels/experiments/mnist_ssl_vae/'
ignore = ['debug-mnist']
results = {}
stats = {}
for expt in os.listdir(rootdir):
    exptdir = os.path.join(rootdir,expt)
    if os.path.isdir(exptdir) and expt not in ignore:
        for d in os.listdir(exptdir):
            if os.path.isdir(os.path.join(exptdir,d)):
                r = getResultsAcrossSeeds(exptdir,d)
                if r != None:
                    if exptdir not in results.keys():
                        results[exptdir] = {}
                    if exptdir not in stats.keys():
                        stats[exptdir] = {}
                    result = results[exptdir]
                    stat = stats[exptdir]
                    results[d] = r
                    stat[d] = getStats(results[d])
                else:
                    print '\n>>>WARNING: no results for %s' % os.path.join(exptdir,d)
            else:
                print '\n>>>WARNING: not a directory: %s' % os.path.join(exptdir,d)
    else:
        print '\n>>>WARNING: not a directory: %s' % exptdir
```

```python
def plotStuff(stat,label,func=min,stepsize=1,color=None,**kwargs):
    n = len(stat['mean'])
    bestval = func(stat['mean'])
    label = '%s: %0.2f' % (label,bestval)
    t = np.arange(1,n+1)*stepsize
    #print stat.keys()
    #print stat['epochs'].shape,stat['mean'].shape,stat['p10'].shape,stat['p90'].shape
    if 'epochs' in stat.keys():
        plt.plot(stat['epochs'],stat['mean'],label=label,**kwargs)
        plt.fill_between(stat['epochs'],stat['p10'],stat['p90'],color=color,alpha=0.2)
    else:
        plt.plot(stat['mean'],label=label,**kwargs)
        plt.fill_between(range(0,n),stat['p10'],stat['p90'],color=color,alpha=0.2)

def plotKey(sorted_stats,key,func=min,stepsize=1):
    for i,(k,v) in enumerate(sorted_stats):
        #print '>>> %s' % k
        if key in v.keys():
            plotStuff(v[key],label=i,func=func,stepsize=stepsize)
    plt.title(key)

def printKeys(stats):
    print stats[stats.keys()[0]].keys()
    
#printKeys(stats)
        
```

```python
def plotExperiment(myexpt,
                   keys = ['train_cost','train_bound','train_accuracy','valid_cost','valid_bound','valid_accuracy'],
                   funcs = [min,min,max,min,min,max],
                   yscales = ['linear','linear','linear','linear','linear','linear'],
                   ylims = [(100,300),(100,300),(0,1),(300,600),(1e2,1e3),(0,1)],
                   locs = ['upper right','upper right','lower right','upper right','upper right','lower right'],
                   stepsizes = [1,1,1,10,10,10]):
    
    #myexpt = '/scratch/jmj/theanomodels/experiments/mnist_ssl_vae/000_batchnorm_maxout-mnist'
    mystat = stats[myexpt]

    print '*'*100
    print myexpt
    print '*'*100

    sorted_stats = sorted(mystat.iteritems(),key=lambda x:x[0])
    for i,(k,v) in enumerate(sorted_stats):
        print '%s:  %s' % (i,k)
    plt.figure(1,figsize=(18,8))
    for i,(key,f,yscale,ylim,loc) in enumerate(zip(keys,funcs,yscales,ylims,locs)):
        #print i,key
        plt.subplot(2,3,i+1)
        plotKey(sorted_stats,key,func=f)
        plt.legend()
        plt.legend(loc=loc,framealpha=0.5)
        plt.yscale(yscale)
        if ylim is not None:
            plt.ylim(ylim)
    plt.show()
```
